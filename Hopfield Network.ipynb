# Understanding Hopfield Networks

Hopfield Networks are a type of recurrent artificial neural network used for associative memory and pattern recognition. They were introduced by John Hopfield in 1982 and are a form of energy-based model, originally derived from the Ising model of ferromagnetism. Hopfield Networks have applications in various fields, including optimization, image recognition, and content-addressable memory.

## Network Structure

A Hopfield Network consists of a set of interconnected neurons, where each neuron can be in one of two states: +1 (active) or -1 (inactive). The neurons are fully connected to each other, meaning that each neuron is connected to every other neuron in the network. The connections between neurons are symmetric, meaning that if neuron $i$ is connected to neuron $j$, then neuron $j$ is also connected to neuron $i$.

## Energy Function

The energy of a Hopfield Network is defined by an energy function, which measures how well the current state of the network matches the stored patterns. The energy function for a Hopfield Network is given by:

$$
E(\mathbf{s}) = -\frac{1}{2}\sum_{i=1}^{N}\sum_{j=1}^{N}w_{ij}s_is_j
$$

Where:
- $E(\mathbf{s})$ is the energy of the network in state $\mathbf{s}$
- $N$ is the number of neurons in the network
- $w_{ij}$ is the weight (synaptic connection strength) between neuron $i$ and neuron $j$
- $s_i$ is the state of neuron $i$, which can be +1 or -1

The energy function represents the negative of the network's Hamiltonian, and it is used to measure the compatibility between the current state of the network and the stored patterns. Lower energy indicates a better match with the stored patterns.

## The Energy Landscape


In Hopfield networks, the energy landscape is a fundamental concept that underlies the network's behavior. The energy of a Hopfield network is a scalar value associated with each state of the network, denoted as E. It is defined by the following equation:

$$
E = -\frac{1}{2} \sum_{i,j} w_{ij} s_i s_j - \sum_{i} \theta_i s_i
$$

where
- $E$ is the total energy of the network,
- $w_{ij} $ is the strength of the connection weight from unit $ j $ to unit $ i $,
- $ s_i $ is the state of unit $ i $, and
- $ \theta_i $ is the threshold of unit $ i $


This energy function is a Lyapunov function, meaning it decreases or stays the same during the network's dynamics. When training a Hopfield network, the goal is to lower the energy of states that the network should "remember." This allows the network to serve as a content addressable memory system.

## States

In the context of a Hopfield network, a "state" refers to the configuration of the network's neurons at a specific time. Each neuron in a Hopfield network can be in one of two states, typically represented as -1 or +1. These states are often used to indicate whether a neuron is "inactive" or "active", respectively. A state of the network is essentially a vector of these neuron states, where each element of the vector corresponds to the state of a specific neuron in the network.

For example, in a Hopfield network with 3 neurons, a state could be represented as [1, -1, 1], indicating that the first neuron is active, the second neuron is inactive, and the third neuron is active.

In the context of Hopfield networks, a state is considered a fixed point if it is a local minimum of the energy function. This means that the state does not change under the network's dynamics, and it represents a stable configuration of the network. When determining if a state is a fixed point in a Hopfield network, it involves evaluating whether the energy of the state remains unchanged during the network's dynamics.

## Basins of Attraction

During the training process, the network is moved to basins of attraction, which are stable states in the energy landscape.  In summary, in the context of a Hopfield network, a "state" refers to the configuration of the network's neurons, with each neuron being in an "inactive" or "active" state. A state is considered a fixed point if it is a local minimum of the energy function, indicating a stable configuration of the network.

These basins of attraction are created by the stored patterns in the network. When a pattern is recalled from the network, the network's energy changes, and the dynamics are attracted toward the stable fixed points corresponding to the stored patterns. The attractors represent local minima in the energy landscape, and the network will always recall one of the stored patterns.

Mathematically, the energy landscape can be visualized with multiple minima of the same depth, with each minimum corresponding to the retrieval of one of the patterns. The attractor basins represent the stability of the network's dynamics, and the deeper the basin, the higher the attraction. The network's dynamics are 'attracted' toward these stable fixed points, making Hopfield networks also known as 'attractor' networks or 'attractor memories'.

In summary, the energy landscape in Hopfield networks is a key concept that governs the network's behavior. During training, the network is moved to basins of attraction, which are stable states in the energy landscape created by the stored patterns. When recalling patterns, the network's dynamics are attracted toward these stable points, allowing for robust and reliable pattern retrieval.

## Energy Decrease with a Small Network

In a 2-node Hopfield network, the energy decreases as the network evolves due to the dynamics governed by the update rule, which is designed to minimize the energy function. Let's consider a specific example with a 2-node network. 

Suppose we have a 2-node Hopfield network with the following weight matrix:

$$
W = \begin{bmatrix}
0 & w \\
w & 0
\end{bmatrix}
$$

where $$w$$ is the weight of the connection between the two nodes.

The energy of the network for a given state $$\mathbf{s} = [s_1, s_2]$$ is given by:

$$
E = -\frac{1}{2} \mathbf{s}^T W \mathbf{s}
$$

For simplicity, let's consider the case where the weight $$w = 1$$. Now, let's examine how the energy changes for different states of the network:

1. When both nodes are in the same state (either both -1 or both +1), the energy is -1/2.
2. When the nodes are in different states (+1 and -1), the energy is 1/2.

As we can see, the energy is lower when the nodes are in the same state, and higher when the nodes are in different states. This demonstrates that the energy decreases when the nodes converge to the same state, which is a fundamental property of Hopfield networks.

## Energy Landscape for 2 Nodes

The energy landscape for a 2-node Hopfield network can be visualized as a bowl-shaped surface. This is because the energy function for a 2-node Hopfield network is a quadratic form, which corresponds to a bowl-shaped energy surface. The bowl-shaped surface represents the network's energy levels across its entire configuration space, and the network's dynamics are such that it tends to converge towards the bottom of this bowl, which corresponds to stable states or attractors.

The energy landscape visualization for a 2-node Hopfield network provides an intuitive understanding of how the network's energy decreases as it converges towards stable states, and how the energy function acts as a Lyapunov function, guiding the network's dynamics towards these stable configurations.

## Convergence of Nodes to the Same State

The convergence of nodes to the same state in a Hopfield network can be demonstrated through a simple example. Consider a 2-node Hopfield network with the following weight matrix:

$$
W = \begin{bmatrix}
0 & 1 \\
1 & 0
\end{bmatrix}
$$

Let's initialize the network with the following initial state: $$\mathbf{s_0} = [1, -1]$$

We can then update the state of the network using the following iterative rule until the network converges to a stable state:

1. Update the state of node 1: $$s_1 = sign(W_{11} \cdot s_1 + W_{12} \cdot s_2) = sign(0 \cdot 1 + 1 \cdot (-1)) = -1$$
2. Update the state of node 2: $$s_2 = sign(W_{21} \cdot s_1 + W_{22} \cdot s_2) = sign(1 \cdot 1 + 0 \cdot (-1)) = 1$$

After the first iteration, the updated state of the network is: $$\mathbf{s_1} = [-1, 1]$$

We continue this iterative process until the network converges to a stable state. In this example, it can be observed that the states of the nodes are converging to the same values, which demonstrates the convergence property of the Hopfield network.

## Training a Hopfield Network

Hopfield Networks are trained to store patterns by adjusting the synaptic weights. The training process is a one-time setup where the network learns to associate patterns with specific energy minima.

The steps are as follows:

1. **Initialization**: Initialize the network with a specific number of neurons.
2. **Define Patterns**: Choose the binary patterns that the network should memorize.
3. **Update Weights**: Adjust the synaptic weights based on the chosen patterns using a learning rule like the Hebbian rule.
4. **Set Diagonal Elements to Zero**: Ensure that neurons do not have self-connections.
5. **State Updating (Dynamics)**: This is the missing step in your notebook where the network's states are iteratively updated based on the current states and the trained weights.
6. **Observation and Analysis**: Visualize and analyze the final states of the network.

### Step 5: State Updating

- After training the network (updating the weights and setting the diagonal elements to zero), the next step is to update the states of the network.
- The state of each neuron is updated based on the weighted sum of its inputs. The update can be synchronous (all neurons at once) or asynchronous (one at a time).
- The update rule for a neuron's state $ s_i $ is typically given by:
  $$
  s_i^{(new)} = \text{sign}\left(\sum_{j} w_{ij} s_j^{(old)} - \theta_i\right)
  $$
  where $ \theta_i $ is the threshold for the ith neuron, and $ w_{ij} $ are the synaptic weights.
- Iterate this process until the network reaches a stable state (an attractor), where further updates do not change the state.

This step is crucial for the network to demonstrate its recall ability, converging to one of the memorized patterns when given an initial state or a corrupted version of a pattern.

```python
# Calculate the weighted sum of inputs for each neuron
weighted_sum = np.dot(hopfield_net.weights, hopfield_net.state_tensor)

# Update states based on the weighted sum and threshold
new_states = np.sign(weighted_sum - threshold)

# Ensuring that the sign function outputs +1 or -1 (not 0)
new_states[new_states == 0] = 1
hopfield_net.state_tensor = torch.tensor(new_states)
```

### Step 6: Repeat for All Patterns
- Repeat steps 3 and 4 for each pattern in the training set.
- The final weight matrix will be the result of accumulating the contributions from all patterns.

Mathematically, the weight update for each pair of neurons (\(i, j\)) is:
$$
w_{ij} = \frac{1}{N} \sum_{\text{pattern}} (s_i \cdot s_j)
$$

After training, the Hopfield Network is ready to perform pattern recognition and recall stored patterns. It can be used to recover a stored pattern given a noisy or partial input.

```python
for pattern in patterns[1:,:]:
    hopfield_net.weights += 1 / 9 * np.outer(pattern, pattern)
    hopfield_net.weights.fill_diagonal_(0)
    
    # Calculate the weighted sum of inputs for each neuron
    weighted_sum = np.dot(hopfield_net.weights, hopfield_net.state_tensor)

    # Update states based on the weighted sum and threshold
    new_states = np.sign(weighted_sum - threshold)

    # Ensuring that the sign function outputs +1 or -1 (not 0)
    new_states[new_states == 0] = 1
    hopfield_net.state_tensor = torch.tensor(new_states)

    draw_hopfield_network(hopfield_net.state_tensor, hopfield_net.weights, edge_color = 'grey', draw_self_loops = False)
#     print('edge weights:\n', hopfield_net.weights)
```
